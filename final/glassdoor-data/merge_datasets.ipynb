```python
import pandas as pd

firmen_db = []
with open('data_firmen.csv','r',encoding='utf8') as f:
    for l in f:
        firmen_db.append(l.split(';'))

firmen_db = pd.DataFrame(firmen_db[1:],columns=firmen_db[0])
firmen_db
```

```python
glassdoor_data = pd.read_csv('clean_glassdoor_data.csv',encoding='utf8')
glassdoor_data
```

```python
import difflib

tmp = 0
names = [n.strip().lower() for n in firmen_db['Offizieller_Name']]
locations = []
for j,name in enumerate(glassdoor_data['company_name']):
    if any(name.strip().lower() in n for n in names):
        tmp += 1
        for i,n in enumerate(names):
            if name.strip().lower() in n:
                
                # firmen_db['Offizieller_Name'][i]
                locations.append((firmen_db['streetAddress'][i]+', '+firmen_db['postalCode'][i]+', '+firmen_db['addressLocality'][i],0))
                break
            # print(name)
    else:
        locations.append(((name.strip().lower() + ' Saarland',glassdoor_data['industry'][j]),1))
        

print(tmp)
```

```python
import googlemaps
from datetime import datetime

with open('apikey.txt') as f:
    api_key = f.readline()
    f.close
    
gmaps = googlemaps.Client(key=api_key)

gmaps_locations = []
for i in range(len(locations)):
    location,ltype = locations[i]
    if ltype==0:
        geocode_result = gmaps.geocode(location)
    else:
        geocode_result = gmaps.find_place(location[0],input_type='textquery')
        if geocode_result['candidates']:
            geocode_result = gmaps.geocode(place_id=geocode_result['candidates'][0]['place_id'])
        else:
            geocode_result = []
        # print(location)
        # print(geocode_result)
        # print('-'*20)
    if len(geocode_result) != 0:
        gmaps_locations.append(geocode_result[0])
    else:
        gmaps_locations.append(None)
```

```python
gmaps_locations[0]
```

```python
street_number = []
street_name = []
town = []
district = []
state = []
postal_code = []
formatted_address = []
lat = []
lng = []
for loc in gmaps_locations:
    if loc: 
        flags = [False,False,False,False,False]
        for comp in loc['address_components']:
            if 'street_number' in comp['types']:
                street_number.append(comp['long_name'])
                flags[0]=True
            if 'route' in comp['types']:
                street_name.append(comp['long_name'])
                flags[1]=True
            if 'locality' in comp['types']:
                town.append(comp['long_name'])
                flags[2]=True
            if 'administrative_area_level_1' in comp['types']:
                state.append(comp['long_name'])
            if 'administrative_area_level_3' in comp['types']:
                district.append(comp['long_name'])
                flags[3]=True
            if 'postal_code' in comp['types']:
                postal_code.append(comp['long_name'])
                flags[4]=True
        if not all(flags):
            for i in range(len(flags)):
                if flags[i]==False and i ==0:
                    street_number.append(None)
                if flags[i]==False and i ==1:
                    street_name.append(None)
                if flags[i]==False and i ==2:
                    town.append(None)
                if flags[i]==False and i ==3:
                    district.append(None)
                if flags[i]==False and i ==4:
                    postal_code.append(None)

            # print(loc)

        formatted_address.append(loc['formatted_address'])
        lat.append(loc['geometry']['location']['lat'])
        lng.append(loc['geometry']['location']['lng'])
            
    else:
        street_number.append(None)
        street_name.append(None)
        town.append(None)
        district.append(None)
        state.append(None)
        postal_code.append(None)
        formatted_address.append(None)
        lat.append(None)
        lng.append(None)


print(len(state))
glassdoor_data['street_number']=street_number
glassdoor_data['street_name']=street_name # route
glassdoor_data['town']=town # locality 
glassdoor_data['district']=district # administrative_area_level_3
glassdoor_data['state']=state # administrative_area_level_1
glassdoor_data['postal_code']=postal_code
glassdoor_data['formatted_address']=formatted_address
glassdoor_data['lat']=lat
glassdoor_data['lng']=lng
```

```python
master_df = pd.read_csv('master_df(new feature).csv',encoding='utf8')
master_df
```

```python
names_conventions = {}
for loc in gmaps_locations:
    if loc: 
        for comp in loc['address_components']:

            if 'locality' in comp['types']:
                names_conventions[comp['long_name']] = comp['short_name']
            if 'administrative_area_level_3' in comp['types']:
                names_conventions[comp['long_name']] = comp['short_name']

names_conventions
```

```python
tmp = 0
for t in glassdoor_data['town']:
    if t:
        if any([t.strip().lower() in name or names_conventions[t].strip().lower() in name for name in [n.strip().lower() for n in master_df['Gemeinde'] if n] ]):
            tmp +=1
        else:
            print(t)
print(tmp)
```

```python
data_to_merge = []
for t in glassdoor_data['town']:
    if t:
        for i,n in enumerate(master_df['Gemeinde']):
            flag = True
            if n:
                if t.strip().lower() in n.strip().lower() or names_conventions[t].strip().lower() in n.strip().lower():
                    data_to_merge.append(master_df.iloc[[i]].values.flatten().tolist())
                    # print(data_to_merge)
                    flag = False
                    break
        if flag:
            data_to_merge.append([None]*len(master_df.iloc[[0]].values.flatten().tolist()))
    else:
        data_to_merge.append([None]*len(master_df.iloc[[0]].values.flatten().tolist()))
print(len(data_to_merge))
```

```python
data_to_merge = pd.DataFrame(data_to_merge,columns=master_df.columns)
data_to_merge
```

```python
final_df = pd.concat([glassdoor_data,data_to_merge], axis=1)
final_df#.drop('Unnamed:')
```

```python
final_df.to_csv('merged_df.csv',encoding='utf8')
```
