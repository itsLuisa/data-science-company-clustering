```python
import requests
from bs4 import BeautifulSoup
import time
from collections import defaultdict
import random
```

```python
r = requests.get("http://www.firmendb.de/deutschland/Saarland.php")
soup = BeautifulSoup(r.content, "html.parser")
landkreise = soup.find_all("a")

urls = list()
for l in landkreise:
    if "Saarland_" in l["href"]:
        print(l)
        urls.append(l["href"])
urls = urls[6:-1]
```

```python
firmen_kennzahlen = list()
zahlen_sb = list()
rest_kennzahlen = list()
übrige = list()
```

```python
for landkreis in urls[50:]:
    print(landkreis)
    for i in range(0,1000):
        # Making a GET request
        url = 'http://www.firmendb.de/deutschland/' + landkreis
        if i == 0:
            r = requests.get(url)
        else:
            newurl = url[:-4] + "_" + str(i) + ".php"
            r = requests.get(newurl)
        print(r.url)
        
        # print status code
        print(r.status_code)
        if r.status_code == 404:
            break

        # Parsing the HTML
        soup = BeautifulSoup(r.content, 'html.parser')

        lines = soup.find_all("a")
        
        for line in lines:
            if "../firmen/" in line["href"]:
                # firmen_kennzahlen.append(line["href"])
                # zahlen_sb.append(line["href"])
                # rest_kennzahlen.append(line["href"])
                übrige.append(line["href"]) # do it in chunks so it will not abort in the middle of scraping
        time.sleep(1)
```

```python
print(len(firmen_kennzahlen + zahlen_sb + rest_kennzahlen + übrige)) #42
alle_firmenkennzahlen = firmen_kennzahlen + zahlen_sb + rest_kennzahlen + übrige
```

```python
firmen_kennzahlen_enumerate = [(i,j) for i,j in enumerate(alle_firmenkennzahlen)]
firmen_kennzahlen_enumerate
```

```python
for n, company in firmen_kennzahlen_enumerate[17396:]:
    # n = n-165
    print(n)
    firmen_daten = dict()
    url = "http://www.firmendb.de/" + str(company)
    try:
        r = requests.get(url)
    except ConnectionResetError:
        print("wait and try again")
        time.sleep(600)
        r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser', from_encoding="utf-8")
    x = soup.find_all("dl")
    for table in x:
        keys = table.find_all("dt")
        values = table.find_all("dd")
        for i,j in zip(keys,values):
            if i.text[:-1].strip() == "Adresse":
                entry = j.find_all("span")
                for e in entry:
                    print(e["itemprop"], "\t", e.text)
                    firmen_daten[e["itemprop"]] = e.text
                continue
            morestuff = j.find("a")
            if morestuff != None:
                print(i.text, "\t", morestuff.text)
                firmen_daten[i.text[:-1].strip()] = morestuff.text
            else:
                morestuff2 = j.find("br")
                if morestuff2 != None:
                    print(i.text, "\t", ', '.join(j.text.splitlines()))
                    firmen_daten[i.text[:-1].strip()] = ", ".join(j.text.splitlines())
                else:
                    print(i.text, "\t", j.text)
                    firmen_daten[i.text[:-1].strip()] = j.text
    firmen_daten["name"] = firmen_daten["name"][:150]
    path = "./firmendaten/" + str(n).zfill(5) + "_" + str(firmen_daten["name"]).replace(" ", "_").replace(":", "_").replace("/", "_").replace(".","_").replace("|","_").replace("*","_") + ".txt"
    print(path)
    with open(path, "w", encoding="utf-8") as f:
        for feature in firmen_daten:
            line = str(feature) + "\t" + str(firmen_daten[feature]) + "\n"
            f.writelines(line)

    if n % 35 == 0:
        time.sleep(15)
    if n % 104 == 0:
        time.sleep(30)

    t = random.randint(1,3)
    time.sleep(t)

    #7465 -- 7650 -> 185
    #7645 -- 7830
```
