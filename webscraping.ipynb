```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("http://www.firmendb.de/deutschland/Saarland.php")
soup = BeautifulSoup(r.content, "html.parser")
# print(soup.prettify)
landkreise = soup.find_all("a")

urls = list()
for l in landkreise:
    if "Saarland_" in l["href"]:
        print(l)
        urls.append(l["href"])
# print(urls[6:-1])
urls = urls[6:-1]
```

```python
firmen_kennzahlen = list()
with open("companies.txt", "w", encoding="utf-8") as f:
    for landkreis in urls:
        for i in range(0,1000):
            # Making a GET request
            url = 'http://www.firmendb.de/deutschland/' + landkreis
            if i == 0:
                r = requests.get(url)
            else:
                newurl = url[:-4] + "_" + str(i) + ".php"
                r = requests.get(newurl)
            # r = requests.get('https://nimble-needles.com/patterns/easy-mens-beanie-knitting-pattern/')
            
            # check status code for response received
            # success code - 200
            # print(r)

            # print content of request
            # print(r.content)

            # print request object
            print(r.url)
            
            # print status code
            print(r.status_code)
            if r.status_code == 404:
                break

            # Parsing the HTML
            soup = BeautifulSoup(r.content, 'html.parser')
            # print(soup.prettify())

            lines = soup.find_all("a")
            
            for line in lines:
                if "../firmen/" in line["href"]:
                    print(line.text)
                    firmen_kennzahlen.append(line["href"])
                    f.writelines(line.text + "\n")
```

```python
rechtsformen = defaultdict(int)
with open("companies.txt", encoding="utf-8") as f:
    for line in f:
        ending = line.strip().split()[-1]
        # print(ending)
        rechtsformen[ending] += 1

print(sorted(rechtsformen.items(), key=lambda x:x[1], reverse=True))
```

```python
firmen_kennzahlen_enumerate = [(i,j) for i,j in enumerate(firmen_kennzahlen)]
firmen_kennzahlen_enumerate
```

```python
for n, company in firmen_kennzahlen_enumerate[2680:]:
    print(n)
    firmen_daten = dict()
    url = "http://www.firmendb.de/" + str(company)
    try:
        r = requests.get(url)
    except ConnectionResetError:
        print("wait and try again")
        time.sleep(600)
        r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser', from_encoding="utf-8")
    x = soup.find_all("dl")
    for table in x:
        keys = table.find_all("dt")
        values = table.find_all("dd")
        for i,j in zip(keys,values):
            if i.text[:-1].strip() == "Adresse":
                entry = j.find_all("span")
                for e in entry:
                    print(e["itemprop"], "\t", e.text)
                    firmen_daten[e["itemprop"]] = e.text
                continue
            morestuff = j.find("a")
            if morestuff != None:
                print(i.text, "\t", morestuff.text)
                firmen_daten[i.text[:-1].strip()] = morestuff.text
            else:
                morestuff2 = j.find("br")
                if morestuff2 != None:
                    print(i.text, "\t", ', '.join(j.text.splitlines()))
                    firmen_daten[i.text[:-1].strip()] = ", ".join(j.text.splitlines())
                else:
                    print(i.text, "\t", j.text)
                    firmen_daten[i.text[:-1].strip()] = j.text
    firmen_daten["name"] = firmen_daten["name"][:150]
    path = "./firmendaten/" + str(n).zfill(5) + "_" + str(firmen_daten["name"]).replace(" ", "_").replace(":", "_").replace("/", "_") + ".txt"
    print(path)
    with open(path, "w", encoding="utf-8") as f:
        for feature in firmen_daten:
            line = str(feature) + "\t" + str(firmen_daten[feature]) + "\n"
            f.writelines(line)

    if n != 0 and n % 13 == 0:
        # t = random.randint(60,120)
        time.sleep(30)
    if n % 47 == 0:
        # t = random.randint(180,240)
        time.sleep(60)
    if n % 104 == 0:
        # t = random.randint(300,360)
        time.sleep(120)
    # t = random.randint(30,60)
    t = random.randint(5,10)
    # at night
    # t = random.randint(10,20)
    time.sleep(t)
```

```python
import glob
```

```python
alle_firmen = defaultdict(dict)
alle_features = set()
n = 0
firmen_files = './firmendaten/*.txt'
for file in glob.glob(firmen_files):
    n += 1
    with open(file, encoding="utf-8") as f:
        print(file)
        for line in f:
            feature, value = line.strip().split("\t")
            # print(feature, value)
            feature = feature.encode('latin1').decode('ISO-8859-10')
            value = value.encode('latin1').decode('ISO-8859-10')
            feature = feature.strip().strip(":").strip().replace(" ", "_")
            alle_features.add(feature)
            alle_firmen[n][feature] = value.replace(";", ",")
print(alle_features)
print(alle_firmen[18])
```
