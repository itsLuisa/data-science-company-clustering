```python
import pdfplumber
import pandas as pd

pdf_path = "C:/Users/gregc/Downloads/SL_Gemeindezahlen_2021.pdf"

```

```python
target_page = 31

with pdfplumber.open(pdf_path) as pdf:
    
    page = pdf.pages[target_page - 1]  # Adjusting for zero-based indexing
    #cropped = page.crop(bounding_box) 
    extracted_tables = page.extract_tables(table_settings={
        "vertical_strategy": "text",
        "horizontal_strategy": "text",

       
    })

df = pd.DataFrame(extracted_tables[1:], columns = extracted_tables[0])
    

df.head()
```
